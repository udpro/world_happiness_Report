# Dystopia Residual

```python
plt.figure(figsize=(8,5))
sns.histplot(traindata['Dystopia Residual'])
plt.title('Dystopia Residual', fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(traindata['Dystopia Residual'])
plt.title('Dystopia Residual', fontsize=25)
plt.show()
```

# Happiness Score

```python
plt.figure(figsize=(8,5))
sns.histplot(traindata['Happiness Score'], kde=True)
plt.title('Happiness Score', fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(traindata['Happiness Score'])
plt.title('Happiness Score', fontsize=25)
plt.show()
```

```python
corrmat = traindata.corr()
sns.color_palette("Paired")
sns.heatmap(corrmat, vmax=.8, square=True, cmap="PiYG", center=0)
```

```python
plt.figure(figsize=(10,4))
sns.countplot(traindata['Region'])
plt.title('Region', fontsize=25)
plt.show()
```

```python
corrmat = traindata[traindata['Region'].str.contains('Africa')].corr()
plt.figure(figsize=(8,5))
ax = plt.axes()
ax.set_title('Africa')
sns.heatmap(corrmat, annot=True, vmax=.8, square=True, cmap="PiYG", center=0,)
```

```python
corrmat=traindata[traindata['Region'].str.contains('Asia')].corr()
plt.figure(figsize=(8,5))
ax=plt.axes()
ax.set_title('Asia')
sns.heatmap(corrmat, annot=True, vmax=8, square=True, cmap="PiYG", center=0, )
```
